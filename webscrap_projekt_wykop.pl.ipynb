# Articles information

Upvote scores

````python
upvotes = []
for tag in soup.find_all("div", class_="dig"):
    number_match = re.search(r"\d+", tag.text.strip())
    if number_match:
        upvotes.append(number_match.group())

print(upvotes)
````

Titles of the articles

````python
titles=[]
for h2 in soup.find_all('h2'):
    anchor = h2.find('a')
    if anchor:  # Check if there's an anchor tag within the h2
        titles.append(anchor.text.strip())
# print the extracted titles
print(titles)
````

Tags

````python

hrefs = []
for a in soup.find_all('a',{'data-v-7df2b179': True}):  # Find all <a> elements, this is
    #because data-v-7df2b179 is for tags from posts but there are other tags on the site as well that we are not interested in
    try:
        href = a.attrs['href']  # Get the value of the 'href' attribute
        tag_match = re.match(r"/tag/([^/]+)$", href)  # Search for the tag name using regex
        if tag_match:  # If a match is found
            tag_name = tag_match.group(1)  # Extract the tag name from the matched group,
            #we use tag_match.group(1) so that regex will only the tag from the html syntax

            hrefs.append(tag_name)  # Append the extracted tag name to the list
    except:
        continue

print(hrefs)
````

Comment counter

````python
# when looking at the parsed html we can see that comments have <a class="comment-counter"
comments = []
for comment in soup.find_all("a", class_="comment-counter"):
    comments_match = re.search(r"\d+", comment.text.strip())
    if comments_match:
        comments.append(comments_match.group())

print(comments)
````

# Let's get to scrapping

````python
website = "https://wykop.pl/hity/roku/strona/1"
chromepath = ChromeDriverManager().install()
service_chrome = Service(executable_path = chromepath)
options_chrome = webdriver.ChromeOptions()
driver_chrome = webdriver.Chrome(service = service_chrome, options = options_chrome) # opens chrome

driver_chrome.maximize_window() # mazimizes browser's window
driver_chrome.get(website) # opens a website

time.sleep(5)

cookies_button_xpath ='/html/body/div[2]/div[1]/div[2]/div/div[2]/button[2]' #relative xpath for accepting cookies
content = driver_chrome.find_element('xpath',cookies_button_xpath) # finds the button for accepting cookies
content.click() # clicks the button

time.sleep(3) #increase if needed, it was enough



try:
    # Wait for the close button of the ad to be visible
    ad_xpath='/html/body/section/div[2]/main/section[2]/section/button'
    content = driver_chrome.find_element('xpath',ad_xpath) # finds the button
    #if the ad apperas
    content.click() # clicks the button
except:
    # If the ad doesn't appear
    pass
time.sleep(5)

#since there is only 20 news per page and we want about a 1000, let's do 50 pages. We need a loop for that
#we use f-string to combine xpath in quotations with function in {} brackets
start = time.time()
all_upvotes = []
all_comments = []
all_hrefs = []
all_titles = []

for scrap_pages in range(1, 51):
    time.sleep(2)

    if scrap_pages <= 5:
        next_page = f'/html/body/section/div[2]/main/section[2]/div/section/section/section/footer/nav/ul/li[{scrap_pages+7}]/a'
    else:
        next_page = f'/html/body/section/div[2]/main/section[2]/div/section/section/section/footer/nav/ul/li[12]/a'

    # construct JavaScript to click on the next page link
    next_page_script = f"document.evaluate(\"{next_page}\", document, null, XPathResult.FIRST_ORDERED_NODE_TYPE, null).singleNodeValue.click();"
    driver_chrome.execute_script(next_page_script)

    # in order to get the new content from new pages, we need to update the page's html every time it goes to the new page
    new_page_source = driver_chrome.page_source

    soup = BeautifulSoup(new_page_source, 'html.parser')

    # now, scrape data from the updated soup object for upvotes
    upvotes = []
    for tag in soup.find_all("div", class_="dig"):
        number_match = re.search(r"\d+", tag.text.strip())
        if number_match:
            upvotes.append(number_match.group())

    # append upvotes data for the current page to the list
    all_upvotes.append(upvotes)

    # now, scrape data from the updated soup object for comments
    comments = []
    for comment in soup.find_all("a", class_="comment-counter"):
        comments_match = re.search(r"\d+", comment.text.strip())
        if comments_match:
            comments.append(comments_match.group())

    # append comments data for the current page to the list
    all_comments.append(comments)

    hrefs = []
    for a in soup.find_all('a',{'data-v-7df2b179': True}):
        #because data-v-7df2b179 is for tags from posts but there are other tags on the site as well that interrupt
        #scraping tags that we want
        try:
            href = a.attrs['href']  # Get the value of the 'href' attribute
            tag_match = re.search(r"/tag/([^/]+)$", href)  # Search for the tag name using regex
            if tag_match:  # If a match is found
                tag_name = tag_match.group(1)  # Extract the tag name from the matched group, używamy tag_match.group(1),

            #bo chcemy dostać tylko tekst który "wzięła" pierwsza grupa w regexie
            #dzieki temu mamy pewnosc ze z syntaxu html wezmie tylko tag
                hrefs.append(tag_name)  # Append the extracted tag name to the list
        except:
            continue

    all_hrefs.append(hrefs)


    titles=[]
    for h2 in soup.find_all('h2'):
        anchor = h2.find('a')
        if anchor:  # Check if there's an anchor tag within the h2
            titles.append(anchor.text.strip())
    all_titles.append(titles)



driver_chrome.close()


end = time.time()
print("how long the scraping took in seconds:",end-start)
````

Let's make a dataframe out of what we got

````python
df=pd.DataFrame({'Upvotes':all_upvotes,'Comments':all_comments,'Titles':all_titles,})
df
````
